```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/admin/Desktop/online.csv")
```

```python
df.head()
```

```python
df["Revenue"] = df["Quantity"] * df["UnitPrice"]
```

```python
df
```

```python
df.info()
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], format = "%Y-%m-%d")
```

```python
df["InvoiceDate"] = df["InvoiceDate"].dt.date
```

```python
df["InvoiceDate"].max()
```

```python
import datetime
TodayDate = max(df["InvoiceDate"]) + datetime.timedelta(days =1)
```

```python
TodayDate
```

```python
df.head()
```

```python
df["Recency"] = TodayDate - df["InvoiceDate"]
```

```python
df.columns
```

```python
df
```

```python
df_rfm = df.groupby("CustomerID").agg({"InvoiceNo": "count",
                                      "Revenue" : "sum",
                                      "Recency" : "min"})
```

```python
df_rfm
```

```python
df_rfm = pd.DataFrame.rename(df_rfm , columns = {"InvoiceNo" : "Frequency"})
```

```python
df_rfm
```

```python
df_rfm["Recency_quartiles"] = pd.qcut(df_rfm["Recency"], q= 4, labels  = [4,3,2,1])
```

```python
df_rfm
```

```python
df_rfm["Recency"] = df_rfm["Recency"].astype("timedelta64[D]")
```

```python
df_rfm.info()
```

```python
df_rfm["Revenue_quartiles"] = pd.qcut(df_rfm["Revenue"], q= 4, labels  = [1,2,3,4])
```

```python
df_rfm
```

```python
df_rfm["Frequency_quartiles"] = pd.qcut(df_rfm["Frequency"], q= 4, labels  = [1,2,3,4])
```

```python
df_rfm
```

```python
df_rfm.info()
```

```python
df_rfm["Recency_quartiles"] = df_rfm["Recency_quartiles"].astype(int)
```

```python
df_rfm["Revenue_quartiles"] = df_rfm["Revenue_quartiles"].astype(int)
```

```python
df_rfm["Frequency_quartiles"] = df_rfm["Frequency_quartiles"].astype(int)
```

```python
df_rfm.info()
```

```python
df_rfm["TotalScore"] = df_rfm["Recency_quartiles"] + df_rfm["Revenue_quartiles"] + df_rfm["Frequency_quartiles"]
```

```python
df_rfm
```

```python
df_rfm.groupby("TotalScore").mean()
```
